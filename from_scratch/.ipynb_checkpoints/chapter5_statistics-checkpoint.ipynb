```python
# or another way, without pandas

def _median_odd(x):
    return sorted(x)[len(x)//2]
def _median_even(x):
    sorting = sorted(x)
    hi_midpoint = len(x) // 2
    return (sorting[hi_midpoint - 1] + sorting[hi_midpoint] ) / 2
```

```python
_median_even(df[0])
```

```python
def median(v):
    return _median_even(v) if len(v) % 2 == 0 else _median_odd(v)
```

```python
median(df[0])
```

```python
def quantile(x, p):
    p_index = int(p * len(x))
    return sorted(x)[p_index]
```

```python
# value of 50th percentile of the given array
quantile(df[0], .50)
```

```python
def mode(x):
    counts = Counter(x)
    max_count = max(counts.values())
    return [x_i for x_i, count in counts.items()
            if count == max_count]
```

```python
# the mode of a given array, i.e. the value that appears most
mode(df[0])
```

**Dispersion**

```python
# dispersion refers to measures of how spread out our data is such as range
# dispersion of 0 would be when min == 
def data_range(x):
    return max(x) - min(x)
```

```python
data_range(df[0])
```

```python
# from chapter4_linear_algebra
def dot(v, w):
    return sum(v_i* w_i for v_i, w_i in zip(v, w))
assert dot([1,2,3], [4,5,6]) == 32  # or 1*4 + 2*5 + 3*6
```

```python
def sum_of_squares(v):
    """returns the sum of the squares of each element in a vector"""
    return dot(v,v)
assert sum_of_squares([1,2,3]) == 14 # or 1*1 + 2*2 + 3*3
```

```python
# now we can attempt to get variance
def de_mean(xs):
    x_bar = statistics.mean(xs)
    return [x - x_bar for x in xs]
```

```python
# how a single variable debaites from its mean
def variance(xs):
    n = len(xs)
    deviations = de_mean(xs)
    return sum_of_squares(deviations) / (n - 1)
```

```python
variance(df[0])
```

```python
def std_dev(xs):
    return math.sqrt(variance(xs))
```

```python
std_dev(df[0])
```

```python
# range of or difference between the 75th and 25th quantiles
def interquartile_range(xs):
    return quantile(xs, 0.75) - quantile(xs, 0.25)
```

```python
interquartile_range(df[0])
```

```python
def covariance(xs, ys):
    return dot(de_mean(xs), de_mean(ys)) / (len(xs ) - 1)
```

```python
# how two or more variables vary in tandem from their means
covariance(df[0], df[1])
```

```python
# correlation, how interrelated, connect two or more variables are, with the std_dev divided out
```

```python
def correlation(xs, ys):
    stdev_x = std_dev(xs)
    stdev_y = std_dev(ys)
    if stdev_x > 0 and stdev_y > 0:
        return covariance(xs, ys) / stdev_x / stdev_y
    else:
        return 0
```

```python
correlation(df[0], df[1])
```

```python
# or put a much simpler way with pandas
# always between -1 and 1, 0 being no relationship and 1 being perfectly correlated
# .25 would be weak positive relationship
# .50 or more would be strong

df.corr()
```
